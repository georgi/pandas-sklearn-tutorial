# Pandas Quick-start

Pandas is a library for performing analysing tabular data in Python. It works similarly to spreadsheets but with a lot more flexibility and power.



```python
import pandas as pd

pd.DataFrame({
    'age': [10, 20, 30],
    'name': ['Anton', 'Berta', 'Charli']
})
```

## How to load data into pandas

Typically you will load data from CSV files or a database.

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
# Each column is also called a panda Series
df.Age.head()
```

## How to select columns


```python
df[["Age", "Survived"]].head()
```

## How to add new columns

```python
df["Age Group"] = pd.qcut(df.Age, q=5)
```

```python
df.head()
```

## How to select rows

In order to analyze a subset of data it's useful to filter rows based on a condition.



```python
df[df.Age < 35].head()
```

```python
df[df.Age < 20].Survived.mean()
```

```python
df[df.Age >= 20].Survived.mean()
```

## Aggregate by category


```python
df[["Sex", "Age"]].groupby("Sex").mean()
```

```python
df[["Sex", "Survived"]].groupby("Sex").mean()
```

```python
df[["Age Group", "Survived"]].groupby("Age Group").mean()
```

```python
df[["Sex", "Age Group", "Survived"]].groupby(["Sex", "Age Group"]).mean()
```

```python
df[["Sex", "Age Group", "Survived"]].groupby(["Sex", "Age Group"]).mean().plot.barh()
```
